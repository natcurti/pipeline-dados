```python
import csv

with open('../data_raw/dados_empresaB.csv') as file_csv: 
    spamreader = csv.DictReader(file_csv)
    csv_data = []
    for row in spamreader:
        csv_data.append(row)
    
    print(csv_data)
```

```python
column_names = json_empresa_A[0].keys()
column_names
```

```python
column_names = list(column_names)
column_names
```

```python
column_names_csv = list(csv_data[0].keys())
column_names_csv
```

```python
key_mapping = { 
    'Nome do Item': 'Nome do Produto',
    'Classificao do Produto': 'Categoria do Produto', 
    'Valor em Reais (R$)': 'Preo do Produto (R$)', 
    'Quantidade em Estoque': 'Quantidade em Estoque',
    'Nome da Loja': 'Filial', 
    'Data da Venda': 'Data da Venda'
}
```

```python
new_data_csv = []

for old_dict in csv_data:
    new_dict = {}
    for old_key, value in old_dict.items():
        new_dict[key_mapping[old_key]] = value
    new_data_csv.append(new_dict)
```

```python
new_dados_csv = [{key_mapping.get(old_key): value for old_key, value in old_dict.items()} for old_dict in csv_data]
new_dados_csv
```

```python
new_data_csv[0]
```

```python
combined_list = []
combined_list.extend(json_empresa_A)
combined_list.extend(new_data_csv)
```

```python
len(combined_list)
```

```python
column_names_save = list(json_empresa_A[0].keys())
column_names_save
```

```python
save_path = '../data_processed/dados_combinados.csv'

with open(save_path, mode='w') as file_to_save: 
    writer = csv.DictWriter(file_to_save, fieldnames=column_names_save)
    writer.writeheader()

    for row in combined_list:
        writer.writerow(row)    
```

```python
columns_table_csv = list(combined_list[-1].keys())
column_names_csv
```

```python
combined_data_table = [columns_table_csv]

for row in combined_list:
    new_row = []
    for column in columns_table_csv:
        new_row.append(row.get(column, 'Indisponivel'))
    combined_data_table.append(new_row)

combined_data_table
```

```python
save_path = '../data_processed/dados_combinados.csv'

with open(save_path, mode='w') as file_to_save: 
    writer = csv.writer(file_to_save)
    writer.writerows(combined_data_table)
```
